```python
import os
import time
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
functions = [{
    'name': 'calculator', 
    'description': 'Evaluates arbitrary math expressions.', 
    'parameters': {
        'type': 'object', 
        'properties': {
            'expression': {
                'type': 'string', 
                'description': 'A string representing a mathematical expression. Example: 3 * 8 + 1', 
                'enum': []
                }
            }
        }, 
    'required': ['expression']
}]
```

```python
response = openai.ChatCompletion.create(
  model="gpt-4",
  messages=[
    {
      "role": "system",
      "content": "You are a bot that helps the user calculate various things. You will be provided with a tool for that. Use it when the user asks you to evaluate an expression.\nBefore using the tool, tell the user 'Calculating...'"
    },
    {
      "role": "user",
      "content": "What's 3*3?"
    }
  ],
  temperature=0.7,
  functions=functions,
  frequency_penalty=0,
  presence_penalty=0,
  stream=True
)

iterator = iter(response)
while True:
    start_time = time.time()
    try:
        chunk = next(iterator)
        print(chunk)
        delta_time = time.time() - start_time
        # delta_text = chunk['choices'][0]
        
        # if 'text' not in delta_text:
        #     continue
        
    except StopIteration:
        break
```
